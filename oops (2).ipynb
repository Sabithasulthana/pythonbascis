**Inheritance**

**single** **inheritance**

single inheritance allow to derivate class into inherit properties of one part

```python
class A:
  def printa(self,num):
    self.num=num
    print('inside class A')
class B(A):
  def printb(self):
    print('inside class B')
obj=B()
obj.printa('hello')
obj.printb()
```

**multiple** **inheritance**:

if a class able to be created  from multiple base clases. this kind of inheritance is known as multiple inheritance.each of the arttributes that are present in the classes of the base has been passed on the classes that is derived from it.

```python
class A:
  def printa(self):
    print('inside class A')
class B:
  def printb(self):
    print('inside class B')
class C(A,B):
  def printc(self):
    print('inside class C')
obj = C()
obj.printa()
obj.printb()
obj.printc()
```

**Multi**  **Level** **Inheritance**

The features that are part of the original class as the class that is derived from it, are passed on to the new class. it is similar to a relationship involving grandparents and children.

```python
class A:
  def printa(self):
    print('inside class A')
class B(A):
  def printb(self):
    print('inside class B')
class C(B):
  def printc(self):
    print('inside class C')
obj =C()
obj.printa()
obj.printb()
obj.printc()
```

**Abstraction**

ABSTRACTION is used in python to hide the internal functionaly of the function from the users. in python abstraction is used to hide the irrelevent data/class in order to reduce the complexicity.





```python
from abc import ABC , abstractmethod
class Mail(ABC):

  @abstractmethod
  def mileage(self):
    pass
  def fun(self):
    print('fun')
class Child(Mail):
   def sum(self):
    print('sum')
   def mileage(self):
    print('mileage')

obj1= Child()
obj1.mileage()


```
