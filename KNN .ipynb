**K** **NEAREST** **NEIGHBOUR** **ALGORITHM**

K nearest neighbour is one of the simplest machine learning algorithms based on supervised learning technique

K NN algorithm assumes the similary b/w the nwe case/ data and available cases and put the new case into the category that is most similar to the available categories

Knn algorithm stores all the available data and classifies new data points based on the similarity. this means when new data appears then it can be easily classsified into a well suite category by using k-nn algorithm.

**K** **ALGORITHM**

1: Select the number k of the neighbours.

2: Calculate the euclidean distsnce of the k number of the neighbours.

3; Take the k nn as per the calculated eucledian distance.

4: Among these k neighbours, count the number of the data points in each category

5; Assign the new data points tothat catagory for which the number of the neighbour is maximum

6: Our model is ready

**EUCLIDEAN** **DISTASNCE** **FORMULA**

$$d=\sqrt{x_2-x_1)^2+(y_2-y_1)^2}$$

Manhatten Formula

$$d = |x_2 - x_1| + |y_2 - y_1|$$

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
x= np.array([[150,50],[155,55],[160,60],[161,59],[158,65]])
y=np.array(['Medium','Medium','Large','Large','Large'])
test_point=np.array([[157,54]])
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
prediction=knn.predict(test_point)
print(f'The predicted targect for Height=157 and Weight =54 is:{prediction[0]}')
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
df=pd.DataFrame(Scaler.fit_transform(df),columns=df.columns)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x= df.drop(['target'],axis=1)
y=df['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
print('shape of the train data',x_train.shape)
print('shape of test data',x_test.shape)
print('shape of y_train',y_train.shape)
print('shape of y_test',y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
#n_neighbors=5
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(f'Train score: {knn.score(x_train,y_train)}')
print(f'Test score:{knn.score(x_test,y_test)}')
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)

```
