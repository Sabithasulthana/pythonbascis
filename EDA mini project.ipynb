# ***AIR*** ***QUALITY*** ***AND*** ***POLLUTION*** ***ASSESSMENT***

This dataset focuses on air quality assessment across various regions. The dataset contains 5000 samples and captures critical environmental and demographic factors that influence pollution levels.

Key Features:

Temperature (°C): Average temperature of the region.
Humidity (%): Relative humidity recorded in the region.
PM2.5 Concentration (µg/m³): Fine particulate matter levels.
PM10 Concentration (µg/m³): Coarse particulate matter levels.
NO2 Concentration (ppb): Nitrogen dioxide levels.
SO2 Concentration (ppb): Sulfur dioxide levels.
CO Concentration (ppm): Carbon monoxide levels.
Proximity to Industrial Areas (km): Distance to the nearest industrial zone.
Population Density (people/km²): Number of people per square kilometer in the region.
Target Variable: Air Quality Levels

Good: Clean air with low pollution levels.
Moderate: Acceptable air quality but with some pollutants present.
Poor: Noticeable pollution that may cause health issues for sensitive groups.
Hazardous: Highly polluted air posing serious health risks to the population.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/archive.zip')
plt.figure(figsize=(8,6))
AirQualitylist=df['Air Quality'].tolist()
Temperaturelist=df['Temperature'].tolist()
plt.plot( AirQualitylist,Temperaturelist,color='c',label= 'Temperature',marker='^')
plt.title('Relation Between Temperature And Airquality ')
plt.xlabel('Air Quality')
plt.ylabel('Temperature')
plt.xticks( AirQualitylist)
plt.yticks([10,20,30,40,50,60,70,80])
plt.legend()
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/archive (12).zip')
df.head(10)
```

```python
import matplotlib.pyplot as plt
pollutants=['PM2.5','PM10','NO2','SO2','CO']
proportions=[df['PM2.5'].mean(),df['PM10'].mean(),df['NO2'].mean(),df['SO2'].mean(),df['CO'].mean()]
plt.figure(figsize=(8,6))
plt.pie(proportions,labels=pollutants,radius=1,startangle=90,autopct='%1.1f%%',explode=(0,0,0,0,1))
plt.show()
```

The graph indicates that particulate matter (PM10 and PM2.5) and nitrogen dioxide (NO2) are the most significant air pollutants, with PM10 being the highest. Sulfur dioxide (SO2) contributes a smaller percentage, and other pollutants make up a minor fraction.

```python
import seaborn as sns
import pandas as pd
sns.lmplot(x='PM2.5',y='PM10',data=df,fit_reg=False)
```

The upward trend of the data points indicates that higher levels of PM2.5 are associated with higher levels of PM10.

There is some scatter around the trend line, which indicates that the relationship is not perfectly linear. This means that for any given level of PM2.5, there can be some variation in the corresponding PM10 levels.
