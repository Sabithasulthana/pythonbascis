*`What`* *is* *overfitting*

When  a model performs very well for training data but has poor performance with test data. it is knowm as overfiting . In this case, the machine learning model learns the details and noise in the training data such that it negatively affectcs the performance of the model on test data. Overatting can happen due tp low bias and high variance

Reason for overfitting

  Data used for training is not cleaned and contains noise in it.

The model has high variance

The size of the trainig datasets used is not enough.

The model is too complex

What is underfitting

When  a model has not learned the pattern in the training data well and is unable to generalize well on the new data. it is known as undefitting . an underfit model has poor performanceon the training data and will the result in unreliable prediction . Underfitting ocurs due to bias and low variance

Reason of underfitting

Data used for training is not cleaned and contains noise in it.

The model has a high bias.

The size of the training datasets used is not enough.

The model is too simple
