# **Support** **Vector** **Machine** **Algorithm**

Support Vector Machine Algorithm (svm)  is on of the most popular learning algorithms, which is used for classification as well as regrassion problems.

**APPLICATIONS**

Classification of microarray gene expression data in the field of the bioinformation to identify cancer or other genetic diseases.

Text categorised such as identification on the language used in documents or the classification of documents but subject matter.

The detection of rate yet important events like combustion engine failiure, security breaches etc.

The goals of the SVM algorithms is to created the best line or decision boundary that can segregate n dimentional space into classes so it easy to put the new data points in the correct category in the future, this best decision boundary is called a hyperplane

SVM used a boundary called a hyperplane to partition  data into groups of similar class values in two and three dimension . if the data points can be seperated perfectly by the stright line or flat surface they are said to be lineraly seperatable.



In two dimention the task of the SVM algorothm is identify a line that seperates two classes.

Thre is more than onechoices of dividing line b/w the groups and circles and squares. Three such possiblities are labeled a,b, and c.

SVM algothrim will choose **MAXIMUM** **MARGIN** **HYPERPLANE**(MMH) that creates the greatest seperation b/w the two classes.

Altthough any of the three lines the circles and squares would correctly classify all the data points. it is likely that the line that leads to the greatest separation will generalize the best to the future data

The maximum margin  will improve the chance that, in spite of random noise, the points will remains on the correct side of the boundary

The **supportive** **vector** are the points from each classes that are the closets to MMH , each classes must have at least supportive one vector, but it is possible to have more than one

Using the supportive vector alone, it is possible to define the MMH , this is a key feature of SVM the supportive vectors provide a very compact wat to store a classification models, even if the number of features is extremely large.

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/diabetes.csv')
df
```

```python
df.info()
```

```python
df.head()
```

```python
x=df.iloc[:,0:8]
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn .svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
test_score=svc.score(x_test,y_test)
test_score
```

```python
train_score=svc.score(x_train,y_train)
train_score
```
