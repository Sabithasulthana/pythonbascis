**LIST** **COMPREHENSION**

list comprehension offers a shorter syntax when you want to create a new list based on the values of an exisiting list.

```python
list1=range(10,50)
even=[]
for i in list1:

 if i%2==0:
  even.append(i)
print(even)
```

```python
list1=range(10,20)
even=(i for i in list1 if i%2==0)
print(list(even))
```
