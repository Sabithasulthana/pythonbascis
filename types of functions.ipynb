**zip** **function**

zip function take iterable (can be zero or more ) makes iterable that aggregate elements based on the iterable passed, and iterable or tuple.
synthax
zip(*iterable)


```python
#[2,3,4][3,6,4][5,7,8]=((2,3,5),(3,6,7),(4,4,8))
```

```python
x=[1,2,3]
y=['sabi','misna','ravi']
z=list(zip(x,y))
print(z)
```
