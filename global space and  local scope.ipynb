**Global** **scope**

Avariable created in the main body of the python code os a global variable and blogs to the global scope.
Global variable are variable from within any scope, global and local.
Avariable created outside of the function is the global and can be used by anyone.


```python
x= 300
def myfunc():
  print(x)
myfunc()
print(x)
#this is the global scope
```

**Global** **keyword**

If you need to create a global variable ,But are the stuck in local scope. you can use the global keyword
The global keyword makes the variable global.

```python
def myfunc():
  x= 250
myfunc()
print(x)
```
