# SET

Set is not ordered sequence,
mutable,
 not allow to duplicate ,members,
iterable, infainate length,
return in curly bracket,
unchangable

```python
name={"sabitha","misna","suhana"}
print(name)
```

```python
print(type(name))
```

```python
name={'sabitha', 'suhana','shifa'}
print(name)
```

```python
name.add('radha')
print(name)
```

```python
name.pop()
```

```python
name.remove("sabitha")
print(name)
```

```python
name.clear()
print(name)
```

**union** **method**

the union () method  retrun a new set with  all itemes from both case

```python
name={'sabitha','suhana','shifa'}
name2={'misna','arifa','fida'}
name3=set.union(name,name2)
print(name3)
```

**difference** **method**

```python
name={"sabitha","misna"}
name2={"arifa","fida","misna"}
name3=set.difference(name,name2)
print(name3)
```

```python
name4=set.intersection(name,name2)
print(name4)
```
